# 例題
　
当たりの出る確率が ${p(0\leq p \leq 1)}$ であるくじ引きがある。くじ引きの当たりの出る確率は常に一定であり、また特定の回の結果は次の回の結果に影響を与えないものとする。 　

このようなくじ引きを５回引いたときに、当たりが k 回出たとする。このようなことが起きる確率を p と k を使って表せ。

赤石雅典. Ｐｙｔｈｏｎでスラスラわかる　ベイズ推論「超」入門 (ＫＳ情報科学専門書) (Japanese Edition) (pp. 23-24). (Function). Kindle Edition. 

考え方のメモ：この場合5回の試行は独立である．さらにくじ引きのあたりはずれは0-1のどちらかで表現できる．ここでは，あたりを1とすることにする．

1回の試行がベルヌーイ試行．これを5回してk回1が出る確率．

確率変数${X}$の分布は，二項分布となる．



$$ P(X=k) = {}_5 \mathrm{C}_k p^k (1-p)^{N-k}$$


実際にはくじ引きでは${p = 1/2}$

```python
from scipy.special import comb # Conbinatorial function

n = 5
x = range(n+1)
#y = [comb(n, k)/2**n for k in x]
y = []
for k in x:
    a = comb(n,k)/(2**n)
    y.append(a) 
plt.bar(x ,y)
plt.title('二項分布N=5の確率質量関数')
```
